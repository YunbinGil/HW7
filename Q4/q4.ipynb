```python
import numpy as np
import pandas as pd

def make_newdf():
    data = pd.read_csv('gender2.csv', encoding='ANSI', index_col='행정구역')
    display(data)
    
    total = data['2022년_계_총인구수']
    male = data["2022년_남_총인구수"]
    female = data["2022년_여_총인구수"]
    data_selected = {'Total':total, 'Male':male, "Female":female}
    
    df = pd.DataFrame(data_selected)
    display(df)

    
def main():
    make_newdf()
    
if __name__ == "__main__":
    main()
```
